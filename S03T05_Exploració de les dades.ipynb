#### Crea columnes noves (velocitat mitjana del vol, si ha arribat tard o no...)

```python
df_flight["vel_mitjana_vol"] = df_flight["Distance"]/(df_flight["AirTime"]/60) 
#crea una columna nueva a partir de distancia y tiempo de vuelo: velocidad media (millas/hora)
df_flight["vel_mitjana_vol"] 
```

```python
df_flight["delayed"] = df_flight["ArrTime"] > df_flight["CRSArrTime"] 
#crea una columna nueva según la condición de vuelo que llega tarde: hora de llegada superior a hora programado de llegada
df_flight["delayed"] 
```

#### Taula de les aerolínies amb més endarreriments acumulats 

```python
df_flight_delayed = df_flight[df_flight["delayed"] == True] 
#se crea una tabla con todas los vuelos que han llegado tarde
df_flight_delayed["UniqueCarrier"].value_counts() 
#número de vuelos que han llegado tarde por aerolínea
```

#### Quins són els vols més llargs?

```python
df_flight_dist = df_flight[["FlightNum", "Distance"]].sort_values(by=["Distance","FlightNum"], ascending=False).astype(int)
#se crea un dataframe ordenados por distancia y número de vuelo en descendente
df_flight_dist
```

```python
df_flight_dist.groupby(["FlightNum"]).aggregate(max).sort_values(by=["Distance"], ascending=False).astype(int)
#se agrupa por vuelos y se toma el valor máximo de cada agregado
```

#### I els més endarrerits?
